```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("healthcare.csv")
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset["bmi"].fillna(dataset["bmi"].mean(),inplace=True)
```

```python
dataset.isnull().any()
```

```python
x=dataset.iloc[:,1:11].values
y=dataset.iloc[:,11].values
```

```python
x
```

```python
y
```

```python
dataset.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([("oh",OneHotEncoder(),[0,4,5,6,9])],remainder="passthrough")
#coverts data into binary form
x=ct.fit_transform(x)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0) #test_size=0.2 is 20% testing data
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #used to initialize
from tensorflow.keras.layers import Dense #used to add layers
```

```python
ann=Sequential()
```

```python
ann.add(Dense(10,kernel_initializer ="random_uniform",activation="relu"))
```

```python
ann.add(Dense(20,kernel_initializer ="random_uniform",activation="relu"))
```

```python
ann.add(Dense(1,kernel_initializer ="random_uniform",activation="sigmoid"))
```

```python
ann.compile("adam","binary_crossentropy",metrics=["accuracy"]) #nick name for batch gradient is adam.
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=50)
```

```python
ypred=ann.predict(x_test)
```
