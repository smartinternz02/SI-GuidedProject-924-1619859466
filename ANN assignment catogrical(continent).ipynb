```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset["name"]=le.fit_transform(dataset["name"])
```

```python
x=dataset.iloc[:,0:3].values
y=dataset.iloc[:,3:4].values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
one=OneHotEncoder()
z=one.fit_transform(y[:,0:1]).toarray()
y=np.delete(y,0,axis=1)
y=np.concatenate((z,y),axis=1)
```

```python
y.shape
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4) #test_size=0.2 is 20% testing data
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #used to initialize
from tensorflow.keras.layers import Dense #used to add layers
```

```python
ann=Sequential()
```

```python
ann.add(Dense(3,kernel_initializer ="random_uniform",activation="relu"))
```

```python
ann.add(Dense(6,kernel_initializer ="random_uniform",activation="relu"))
```

```python
ann.add(Dense(8,kernel_initializer ="random_uniform",activation="softmax"))
```

```python
ann.compile("adam","categorical_crossentropy",metrics=["accuracy"]) #nick name for batch gradient is adam.
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=1000)
```

```python
ypred=ann.predict(x_test)
```
